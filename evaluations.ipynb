```python
def display_metrics_evaluation(path_folder, model_name):
    evaluation_result = pd.read_csv(f"{path_folder}/evaluation_result.csv")
    summary_efficiency = pd.read_csv(f"{path_folder}/summary_efficiency.csv")

    print(f"model name: {model_name}")
    print(f"path folder: {path_folder}")
    display(evaluation_result)
    display(summary_efficiency)
    
```

```python
final_path = []
for sub_folders_a in [name for name in os.listdir(result_folder) if os.path.isdir(os.path.join(result_folder, name))]:
    path = f"{result_folder}/{sub_folders_a}"
    for sub_folders_b in [name for name in os.listdir(path) if os.path.isdir(os.path.join(path, name))]:
        display_metrics_evaluation(f"{result_folder}/{sub_folders_a}/{sub_folders_b}", sub_folders_b)
```
